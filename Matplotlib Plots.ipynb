# Matplotlib Plots

**Importing Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Creating Data**

```python
x = np.linspace(0, 5, 11)
y = x**2
```

## I. Functional Approach to Plotting 

### *I. Creating a Plot*

```python
plt.plot(x, y, 'r--')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Figure Title')
plt.show()
```

### *II. Creating Subplots on The Same Canvas*

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r-', marker='*')
plt.xlabel('X1')
plt.ylabel('Y1')
plt.title('First Plot')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b--', marker='*')
plt.xlabel('X2')
plt.ylabel('Y2')
plt.title('Second Plot')
plt.show()
```

## II. Object-Oriented Approach to Plotting 

### *I. Creating a Plot*

```python
# Creating a figure instance/object (i.e. empty canvas)
fig = plt.figure()

# Adding set of axes to figure
# The way of adding set of axes is passing a list of 4 arguements (left, bottom, width, and height ranging between 0 to 1)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # 10% in from left, 10% up from bottom, and takes 80% of the canvas size each in width and height 
# The values represents where the axes to be placed (usually represents the percentage of the blank canvas its going to take)

# Plotting on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('X Label') 
axes.set_ylabel('Y Label')
axes.set_title('Title')
plt.show()
```

### *II. Creating Inset on The Same Canvas*

```python
# Creating a figure instance/object (i.e. empty canvas)
fig = plt.figure()

# Adding set of axes to the larger/main figure 
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Adding set of axes to insert/inset figure 
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Plotting on the set of axes in the larger figure
axes1.plot(x, y, 'b')
axes1.set_xlabel('Main_Fig_X_label')
axes1.set_ylabel('Main_Fig_Y_label')
axes1.set_title('Main Figure')

# Plotting on the set of axes in the insert/inset figure 
axes2.plot(y, x, 'r')
axes2.set_xlabel('Inset_Fig_X_label')
axes2.set_ylabel('Inset_Fig_Y_label')
axes2.set_title('Inset Figure')

plt.show()
```

### *III. Creating Subplots on The Same Canvas*

**Iteration Method**

*1. Basic Iteration: This applies the same plot and labels to every subplot in axes.*

```python
# 1. Basic Iteration

# Creating empty canvas of 2 subplots (nrows, ncols)
fig, axes = plt.subplots(nrows=1, ncols=2) 

# This creates a figure object and adds the set of axes to the figure with the same line of code using tuple unpacking
# If we simply print 'axes', it's is actually an array of matplotlib axes 
# Its essentially just a list of these axes objects meaning we can actually iterate through it.

# Iterating through axes objects
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
    
plt.tight_layout()
plt.show()
```

*2. Indexed Iteration with enumerate(): This applies different data, colors, and labels depending on the subplot's position.*

```python
# 2. Indexed Iteration with enumerate()

# Creating empty canvas of 2 subplots (nrows, ncols)
fig, axes = plt.subplots(nrows=1, ncols=2) 

# Iterating through axes objects
for i, ax in enumerate(axes): # Here, enumerate() simply means "count while looping".
    if i == 0:
        ax.plot(x, y, 'b')
        ax.set_xlabel('First Plot_X_Label')
        ax.set_ylabel('First Plot_Y_Label')
        ax.set_title('First Plot')
    else:
        ax.plot(y, x, 'r')
        ax.set_xlabel('Second Plot_X_Label')
        ax.set_ylabel('Second Plot_Y_Label')
        ax.set_title('Second Plot')

plt.tight_layout()
plt.show()
```

*Indexing Method*

```python
# Creating empty canvas of 2 subplots (nrows, ncols)
fig, axes = plt.subplots(nrows=1, ncols=2)

# Indexing through axes objects
axes[0].plot(x, y, 'b')
axes[0].set_xlabel('First Plot_X_Label')
axes[0].set_ylabel('First Plot_Y_Label')
axes[0].set_title('First Plot')

axes[1].plot(y, y, 'r')
axes[1].set_xlabel('Second Plot_X_Label')
axes[1].set_ylabel('Second Plot_Y_Label')
axes[1].set_title('Second Plot')

plt.tight_layout()
plt.show()
```
